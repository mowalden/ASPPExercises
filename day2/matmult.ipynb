Program to multiply two matrices using nested loops

```python
import random
%load_ext line_profiler
```

```python
N = 250
```

NxN matrix

```python
def create_matrix(N):
    X = []
    for i in range(N):
        X.append([random.randint(0,100) for r in range(N)])
    return X
```

Nx(N+1) matrix

```python
def create_matrix_nonsquare(N):
    Y = []
    for i in range(N):
        Y.append([random.randint(0,100) for r in range(N+1)])
    return Y
```

result is Nx(N+1)

```python
def matmul(X,Y,N):
    result = []
    for i in range(N):
        result.append([0] * (N+1))

    # iterate through rows of X
    for i in range(len(X)):
        # iterate through columns of Y
        for j in range(len(Y[0])):
            # iterate through rows of Y
            for k in range(len(Y)):
                result[i][j] += X[i][k] * Y[k][j]
    return result
```

Test Matrix creation

```python
%lprun -f create_matrix create_matrix(N)
```

```python
%lprun -f create_matrix_nonsquare create_matrix_nonsquare(N)
```

We can see that the matrix creations are slow, but they still take less than 1s each.

Test matrix multiplication

```python
X = create_matrix(N)
Y = create_matrix_nonsquare(N)
%lprun -f matmul matmul(X,Y,N)
```

One can see that the matrix multiplication takes up most of the time. One should stark by optimizing line 9-12.
